```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("Cleaned_Dataset.xlsx")
```

```python
df["Product"].value_counts().plot(
    kind="bar",
    figsize=(8,5)
)

plt.title("Most Ordered Products")
plt.xlabel("Product")
plt.ylabel("Number of Orders")
plt.show()
```

```python
df["PaymentMethod"].value_counts().plot(
    kind="bar",
    figsize=(6,4)
)

plt.title("Payment Method Distribution")
plt.xlabel("Payment Method")
plt.ylabel("Orders")
plt.show()
```

```python
df["OrderStatus"].value_counts().plot(
    kind="bar",
    figsize=(6,4)
)

plt.title("Orders by Status")
plt.xlabel("Order Status")
plt.ylabel("Orders")
plt.show()
```

```python
sales = df.groupby("Product")["TotalPrice"].sum()

sales.plot(
    kind="bar",
    figsize=(8,5)
)

plt.title("Total Sales by Product")
plt.xlabel("Product")
plt.ylabel("Revenue")
plt.show()
```

```python
plt.figure(figsize=(7,5))

plt.hist(df["Quantity"], bins=10)

plt.title("Distribution of Quantity Ordered")
plt.xlabel("Quantity")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(7,5))

plt.hist(df["TotalPrice"], bins=20)

plt.title("Distribution of Order Value")
plt.xlabel("Total Price")
plt.ylabel("Orders")

plt.show()
```

```python
df["Month"] = df["Date"].dt.to_period("M")

monthly = df.groupby("Month")["TotalPrice"].sum()

monthly.plot(
    figsize=(10,5),
    marker="o"
)

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.show()
```

```python
plt.figure(figsize=(7,5))

plt.scatter(
    df["Quantity"],
    df["TotalPrice"]
)

plt.title("Quantity vs Total Price")
plt.xlabel("Quantity")
plt.ylabel("Total Price")

plt.show()
```

```python
plt.figure(figsize=(6,5))

plt.boxplot(df["TotalPrice"])

plt.title("Outliers in Total Price")

plt.show()
```
