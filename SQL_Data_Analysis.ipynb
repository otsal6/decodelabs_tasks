```python
import pandas as pd
import sqlite3
```

```python
df = pd.read_excel("Cleaned_Dataset.xlsx")
```

```python
conn = sqlite3.connect("orders.db")
```

```python
df.to_sql("Orders", conn, if_exists="replace", index=False)
```

```python
pd.read_sql_query("""
SELECT *
FROM Orders;
""", conn)
```

Selected Columns

```python
pd.read_sql_query("""
SELECT OrderID, Product, TotalPrice
FROM Orders;
""", conn)
```

WHERE

```python
pd.read_sql_query("""
SELECT *
FROM Orders
WHERE PaymentMethod='Online';
""", conn)
```

ORDER BY

```python
pd.read_sql_query("""
SELECT *
FROM Orders
ORDER BY TotalPrice DESC;
""", conn)
```

```python
pd.read_sql_query("""
SELECT COUNT(*) AS TotalOrders
FROM Orders;
""", conn)
```

AVG

```python
pd.read_sql_query("""
SELECT AVG(TotalPrice) AS AverageOrderValue
FROM Orders;
""", conn)
```

SUM

```python
pd.read_sql_query("""
SELECT SUM(TotalPrice) AS TotalRevenue
FROM Orders;
""", conn)
```

GROUP BY

```python
pd.read_sql_query("""
SELECT Product,
COUNT(*) AS Orders
FROM Orders
GROUP BY Product;
""", conn)
```

GROUP BY + SUM

```python
pd.read_sql_query("""
SELECT Product,
SUM(TotalPrice) AS Sales
FROM Orders
GROUP BY Product
ORDER BY Sales DESC;
""", conn)
```

GROUP BY + AVG

```python
pd.read_sql_query("""
SELECT PaymentMethod,
AVG(TotalPrice) AS AvgOrder
FROM Orders
GROUP BY PaymentMethod;
""", conn)
```

```python
conn.close()
```
